```python
#5.Generate a synthetic dataset with correlated features.Visualize the correlation matrix of a dataset using a heatmap.
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate synthetic dataset with correlated features
np.random.seed(0)
# Generate correlated features
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # Covariance matrix indicating correlation between features
data = np.random.multivariate_normal(mean, cov, 1000)

# Visualize correlation matrix using heatmap
correlation_matrix = np.corrcoef(data, rowvar=False)
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Add title
plt.title('Correlation Matrix Heatmap')

# Display the plot
plt.show()
```

```python
#4.using the given x and y data generate a 3D surface plot to visualize the function
"""
 x= np.linespace(-5,5,100)
 y=np.linespace(-5,5,100)
 z=np.sin(np.sqrt(x**2 + y ** 2))

 data = {
     'X' : x.flatten()
     'Y' :y.flatten()
     'Z':z.flatten()
 }
 df = pd.DataFrame(data)
 """

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Given data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

# Display the plot
plt.show()
```

```python
#1.create a bokeh plot displiying a sine wave.Set x-value from 0 to 10 and y-value as the sine of x.
from bokeh.plotting import figure, show
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='sin(X)')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="sin(x)", line_width=2)

# Show the plot
show(p)
```

```python
#2.create a bokeh scatter plot using randomly generated x and y values use different size and color for the makers based on the sizes and color column
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate random data
np.random.seed(0)
num_points = 100
x = np.random.rand(num_points) * 10  # Random x values
y = np.random.rand(num_points) * 10  # Random y values
sizes = np.random.randint(5, 20, num_points)  # Random sizes for markers
colors = np.random.rand(num_points) * 100  # Random colors for markers

# Create a ColumnDataSource with data
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create a new plot with a title and axis labels
p = figure(title="Scatter Plot", x_axis_label='X', y_axis_label='Y')

# Add scatter markers
p.scatter(x='x', y='y', size='sizes', color='colors', source=source, fill_alpha=0.6)

# Show the plot
show(p)
```

```python
#3.generate a bokeh bar chart representing the counts of different fruits using the following dataset.

  #fruits = ['Apples','Oranges','Bananas','Pears']
  #counts  =[20,25,30,35]

from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource

# Given dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource with data
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a new plot with a title and axis labels
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts')

# Add bars
p.vbar(x='fruits', top='counts', width=0.5, color='skyblue', source=source)

# Show the plot
show(p)
```

```python
#4.craete a bokeh histogram to visualize the distribution of the given data
    #data_hist =np.random.randn(1000)
   # hist , edges = np.histogarm(data_hist,bins=30)

from bokeh.plotting import figure, show
import numpy as np

# Given data
data_hist = np.random.randn(1000)

# Compute histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a new plot with a title and axis labels
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

# Add a quad glyph
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="skyblue", line_color="white")

# Show the plot
show(p)
```

```python
#5.Create a bokeh heatmap using the providd dataset
"""
  data-heatmap = np.random.rand(10,10)
  x =np.linspace(0,1,10)
  y=np.linspace(0,1,10)
  xx , yy =np.meshgrid(x,y)
  """

from bokeh.plotting import figure, show
from bokeh.transform import linear_cmap
from bokeh.util.hex import hexbin
import numpy as np

# Given dataset
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a new plot with a title and axis labels
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y')

# Convert data to 1D arrays
xx_flat = xx.flatten()
yy_flat = yy.flatten()
data_flat = data_heatmap.flatten()

# Create a heatmap
p.rect(x=xx_flat, y=yy_flat, width=1/10, height=1/10, fill_color=linear_cmap('data_flat', 'Viridis256', 0, 1), line_color=None, source={'data_flat': data_flat})

# Show the plot
show(p)

```
